```python
dados = CustomeDataSet(traindata_url, criterio='Data')
dados[0]
```

```python
data = {'index':[],
       'material':[],
        'espessura':[],
       'num_parafusos':[],
        'bitola':[],
        'fs':[]
       }
```

```python
for i in range(len(dados)):
    temp = [i, *dados[i]]
    if dados[i][-2] == 22 and dados[i][0] == 'A36':
        for chave,valor in zip(data.keys(), temp):
            data[chave].append(valor)
```

```python
df = pd.DataFrame(data)
df
```

```python
saida = CustomeDataSet(traindata_url, criterio='Image')
saida[406]
```

```python
[saida[i][1] for i in index]
```

```python
import os
# Escreve o arquivo com separador "#;#"
with open("data_train.csv", "w", encoding="utf-8") as f:
    for i in index:
        caption, img_uuid  = saida[i]
        
        for el in ['top','lateral','frontal']:
            img_target = os.path.join('data','data_train','img',el, f'{img_uuid}.png')
            img_mask = os.path.join('data','data_train','mask',el, f'{img_uuid}.png')
            img_base = os.path.join('data','data_train','base',el, f'{img_uuid}.png')
            
            linha = "#;#".join([
                img_base.replace('\\', '/'),
                img_mask.replace('\\', '/'),
                '',
                img_target.replace('\\', '/'),
                caption,
                'local_editing'
            ])
            f.write(linha + "\n")

print("Arquivo 'data/train.csv' salvo com sucesso!")
```
